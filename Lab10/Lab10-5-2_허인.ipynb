## define loss_tracker

```python
def loss_tracker(loss_plot, loss_value, num):
    '''num, loss_value, are Tensor'''
    vis.line(X=num, Y=loss_value, win = loss_plot, update='append')
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

torch.manual_seed(777)
if device =='cuda':
    torch.cuda.manual_seed_all(777)
```

```python
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

trainset = torchvision.datasets.CIFAR10(root='./cifar10', train=True,
                                        download=True, 
                                        transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=512,
                                          shuffle=True, num_workers=0)

testset = torchvision.datasets.CIFAR10(root='./cifar10', train=False,
                                       download=True, 
                                       transform=transform)

testloader = torch.utils.data.DataLoader(testset, batch_size=4,
                                         shuffle=False, num_workers=0)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()

# random training images
dataiter = iter(trainloader)
images, labels = dataiter.next()
vis.images(images/2 + 0.5)

# labels
print(' '.join('%5s' % classes[labels[j]] for j in range(4)))
```

## make VGG16 using vgg.py

```python
import torchvision.models.vgg as vgg
```

```python
# VGG Network: Convolution Layer 쌓을 내용 선언 (make_layers)
cfg = [32,32,'M', 64,64,128,128,128,'M',256,256,256,512,512,512,'M'] 
#13 + 3 =vgg16
```

```python
# VGG Network building
class VGG(nn.Module):
    def __init__(self, features, num_classes=1000, init_weights=True):
        super(VGG, self).__init__()
        self.features = features #parameter -> self.features
        self.avgpool = nn.AdaptiveAvgPool2d((7, 7))

        # Classifier: VGG의 Classifier는 fully connected layer 3개로 이루어져 있다
        self.classifier = nn.Sequential(
            nn.Linear(512 * 4 * 4, 4096), #FC Layer 1
            nn.ReLU(True),
            nn.Dropout(),
            nn.Linear(4096, 4096), #FC Layer 2
            nn.ReLU(True),
            nn.Dropout(),
            nn.Linear(4096, num_classes),) #FC Layer 3
        
        #weight initialize
        if init_weights:
            self._initialize_weights()
    
    # Forward
    def forward(self, x):
        x = self.features(x)
        #x = self.avgpool(x)
        x = x.view(x.size(0), -1)
        x = self.classifier(x)
        return x

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None: nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, 0, 0.01)
                nn.init.constant_(m.bias, 0)
```

```python
#VGG Network -> model
vgg16= VGG(vgg.make_layers(cfg),10,True).to(device)
```

```python
# VGG Model -> test
a=torch.Tensor(1,3,32,32).to(device)
out = vgg16(a)
print(out)
```

```python
# VGG Network: Loss, Optimizer, Lr
criterion = nn.CrossEntropyLoss().to(device)
optimizer = torch.optim.SGD(vgg16.parameters(), lr = 0.005, momentum=0.9)

# lr을을 고정하지지 않고 학습에에 따라라 lr을을 조금씩씩 감소시키며며 학습하는는 방
lr_sche = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.9)
#5번의의 step마다다 optimizer의 lr에 0.9를 곱해준
```

## plot

```python
loss_plt = vis.line(Y=torch.Tensor(1).zero_(), opts=dict(title='loss_tracker', legend=['loss'], showlegend=True))
```

## training

```python
print(len(trainloader))
epochs = 50

for epoch in range(50):  # loop over the dataset multiple times
    running_loss = 0
    lr_sche.step() #epoch이이 5번번 돌때마다 lr이 0.9배씩 줄어든다.

    for i, data in enumerate(trainloader, 0):
        # inputs
        inputs, labels = data
        inputs = inputs.to(device)
        labels = labels.to(device)

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = vgg16(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # statistics
        running_loss += loss.item()
        if i % 30 == 29:    # print every 30 mini-batches
            loss_tracker(loss_plt, torch.Tensor([running_loss/30]), 
                         torch.Tensor([i + epoch*len(trainloader)]))
            print('[%d, %5d] loss: %.3f' %
                  (epoch + 1, i + 1, running_loss / 30))
            running_loss = 0
        
print('Training Finished')
```

```python
#VGG Network: visualization -> Test data Checking
dataiter = iter(testloader)
images, labels = dataiter.next()

# images (GT image)
imshow(torchvision.utils.make_grid(images))
print('GroundTruth: ', ' '.join('%5s' % classes[labels[j]] for j in range(4)))
```

```python
# Compare result of GT & outputs
outputs = vgg16(images.to(device))
_, predicted = torch.max(outputs, 1)
print('Predicted: ', ' '.join('%5s' % classes[predicted[j]] for j in range(4)))
```

```python
# VGG Network: Test Set -> Model Evaluation
correct = 0
total = 0

with torch.no_grad():
    for data in testloader:
        images, labels = data
        images = images.to(device)
        labels = labels.to(device)
        outputs = vgg16(images)
        
        _, predicted = torch.max(outputs.data, 1)
        
        total += labels.size(0)      
        correct += (predicted == labels).sum().item()

print('Accuracy of network on 10000(일만개) test images: %d %%' % (100 * correct / total))
```
